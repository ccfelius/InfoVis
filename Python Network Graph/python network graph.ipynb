```python
# start with the visualizations

import plotly.offline as py
import plotly.graph_objects as go
import networkx as nx
```

```python
# select year
data = copy.deepcopy(year_2017)

sources = data['SOURCE_SUBREDDIT'].tolist()
targets = data['TARGET_SUBREDDIT'].tolist()
nodes = list(set(sources + targets))
```

```python
# create the datastructure, which is a dictionary in a dictionary

import copy

nodes_copy = copy.deepcopy(nodes)
edges = {}
for node in nodes:
    nodes_copy.remove(node)
    temp_dic = {}
    
    for subreddit in nodes_copy:
        temp_dic[subreddit] = 0
    edges[node] = temp_dic
    nodes_copy.append(node)
```

```python
# select the year we make the graph for
data = data.reset_index()
data = data.drop(columns = 'index')
# data.iloc[0]['SOURCE_SUBREDDIT']
for i in range(len(data)):
    source = data['SOURCE_SUBREDDIT'].iloc[i]
    target = data['TARGET_SUBREDDIT'].iloc[i]
    
    edges[source][target] += 1
    edges[target][source] += 1
```

```python
# count the number of times each node appears as a source or target, put it in a dictionary
nodes_dic = {}
count = 0    

for node in nodes:
    for i in range(len(data)):
        source = data['SOURCE_SUBREDDIT'].iloc[i]
        target = data['TARGET_SUBREDDIT'].iloc[i]
        
        if (source==node) or (target==node):
            count += 1

    nodes_dic[node] = count
    count = 0
```

```python
nodes_dic
```

```python
# initialize graph

graph = nx.Graph()
```

```python
edges
```

```python
# Add node for each node
for node in nodes_dic.keys():
    if nodes_dic[node] > 0:
        graph.add_node(node, size = nodes_dic[node])
        
# For each co-appearance between two characters, add an edge
for edge in edges.keys():
    for edge2 in edges[edge].keys():
        
        # Only add edge if the count is positive
        if edges[edge][edge2] > 0:
            graph.add_edge(edge, edge2, weight = edges[edge][edge2])
```

```python
# choose a layout for the graph
pos_ = nx.spring_layout(graph)
```

```python
# function to create an edge between node x and node y, with a given text and width
def make_edge(x, y, text, width):
    return  go.Scatter(x         = x,
                       y         = y,
                       line      = dict(width = width,
                                   color = 'cornflowerblue'),
                       hoverinfo = 'text',
                       text      = ([text]),
                       mode      = 'lines')
```

```python
# For each edge, make an edge_trace, append to list
edge_trace = []
for edge in graph.edges():
    
    if graph.edges()[edge]['weight'] > 0:
        char_1 = edge[0]
        char_2 = edge[1]
        x0, y0 = pos_[char_1]
        x1, y1 = pos_[char_2]
        text  = char_1 + '--' + char_2 + ': ' + str(graph.edges()[edge]['weight'])
        
        trace = make_edge([x0, x1, None], [y0, y1, None], text, 
                           width = 0.3*graph.edges()[edge]['weight']**0.05)
        edge_trace.append(trace)
```

```python
# Make a node trace
node_trace = go.Scatter(x         = [],
                        y         = [],
                        text      = [],
                        textposition = "top center",
                        textfont_size = 10,
                        mode      = 'markers+text',
                        hoverinfo = 'none',
                        marker    = dict(color = [],
                                         size  = [],
                                         line  = None))
# For each node in graph, get the position and size and add to the node_trace
for node in graph.nodes():
    x, y = pos_[node]
    node_trace['x'] += tuple([x])
    node_trace['y'] += tuple([y])
    node_trace['marker']['color'] += tuple(['cornflowerblue'])
    node_trace['marker']['size'] += tuple([0.07*graph.nodes()[node]['size']])
    node_trace['text'] += tuple(['<b>' + node + '</b>'])
```

```python
# Customize layout
layout = go.Layout(
    paper_bgcolor='rgba(255,255,255,255)', # transparent background
    plot_bgcolor='rgba(255,255,255,255)', # transparent 2nd background
    xaxis =  {'showgrid': False, 'zeroline': False}, # no gridlines
    yaxis = {'showgrid': False, 'zeroline': False}, # no gridlines
    title = '2017'
)
# Create figure
fig = go.Figure(layout = layout)
# Add all edge traces
for trace in edge_trace:
    fig.add_trace(trace)
# Add node trace
fig.add_trace(node_trace)
# Remove legend
fig.update_layout(showlegend = False)
# Remove tick labels
fig.update_xaxes(showticklabels = False)
fig.update_yaxes(showticklabels = False)
# Show figure
fig.show()
```
