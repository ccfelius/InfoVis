```python
import csv
political_sub_list = []
with open('political_subs.txt', 'r') as fd:
    reader = csv.reader(fd)
    for row in reader:
        for element in row:
            political_sub_list.append(element)
```

```python
# filter out the unnecesarry characters

for i in range(len(political_sub_list)):
#     political_sub_list[i] = political_sub_list[i].replace("\r\',' '')
    political_sub_list[i] = political_sub_list[i].replace("/r/", "")
    political_sub_list[i] = political_sub_list[i].replace(" ", "")
    political_sub_list[i] = political_sub_list[i].replace("'", "")
    political_sub_list[i] = political_sub_list[i].replace('"', "").lower()
```

```python
 filtered = df.loc[df['SOURCE_SUBREDDIT'].isin(political_sub_list)]
```

```python
filtered = filtered.loc[filtered['TARGET_SUBREDDIT'].isin(political_sub_list)]
```

```python
filtered['TIMESTAMP']= pd.to_datetime(df['TIMESTAMP'])
```

```python
filtered = filtered.sort_values(by="TIMESTAMP")
```
